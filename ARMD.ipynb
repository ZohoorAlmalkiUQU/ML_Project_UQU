## cultures_vitals.parquet file preparing

Processing for `cultures_vitals.parquet`:
My `cultures_vitals.parquet` not have a timestamp column at all:
That means this file (`cultures_vitals.parquet`) already contains aggregated features (e.g., `median`/`first`/`last` only) — not raw time-stamped vitals. So In this case i will:
- Keep cols: `median`, `first`, and `last` values for each vital sign.
- Compute the Delta features: the change between `first` and `last` (e.g., `delta_heartrate`), capturing the trend over time.

The resulting dataset will contain one row per patient with informative, time-aware features that can help predict antibiotic resistance.

````python
sample_folder = 'new_sample_one/'

vitals = pd.read_parquet(sample_folder + "cultures_vitals.parquet")

print(vitals.columns)
print(vitals.shape)
````

````python
import pandas as pd

vitals_cols = [
    'anon_id', 'pat_enc_csn_id_coded', 'order_proc_id_coded',
    'median_heartrate', 'first_heartrate', 'last_heartrate',
    'median_resprate', 'first_resprate', 'last_resprate',
    'median_temp', 'first_temp', 'last_temp',
    'median_sysbp', 'first_sysbp', 'last_sysbp',
    'median_diasbp', 'first_diasbp', 'last_diasbp'
]

vitals[vitals_cols[3:]] = vitals[vitals_cols[3:]].apply(pd.to_numeric, errors='coerce')

vitals = vitals[vitals_cols]

vital_signs = ['heartrate', 'resprate', 'temp', 'sysbp', 'diasbp']

for v in vital_signs:
    first_col = f'first_{v}'
    last_col = f'last_{v}'
    delta_col = f'delta_{v}'
    vitals[delta_col] = vitals[last_col] - vitals[first_col]
````

````python
vitals.info()
````

My vitals data is sparse — for many patients, the first and/or last readings weren't available in the 7-day window.

Using delta features alone may drastically reduce my sample size during modeling or training.

So i will keep rows where i have at least one usable delta - and avoid dropping too much data.

````python
vital_signs = ['heartrate', 'resprate', 'temp', 'sysbp', 'diasbp']

conditions = [
    vitals[f'first_{v}'].notna() & vitals[f'last_{v}'].notna()
    for v in vital_signs
]

condition = conditions[0]
for cond in conditions[1:]:
    condition |= cond

vitals = vitals[condition].copy()

drop_cols = [f'first_{v}' for v in vital_signs] + [f'last_{v}' for v in vital_signs]
vitals.drop(columns=drop_cols, inplace=True)

print(vitals.shape)
````


I will select a meaningful subset of labs:

- `Median` values (central tendency)
- Delta (`last` - `first`) when both are present

Convert object columns to numeric
Calculate delta values like we did for vitals
Drop first_ and last_, keep only delta_ and median_

- I will drop those cols because data `first_*` and `last_*` values are often sparse (e.g., `first_procalcitonin` has only 12 values) + reducing redundancy and dimensionalities + preventing data leakage.

````python
import pandas as pd

sample_folder = 'new_sample_one/'
labs = pd.read_parquet(sample_folder + "cultures_labs.parquet")

labs_cols_base = [
    'anon_id', 'pat_enc_csn_id_coded', 'order_proc_id_coded'
]

labs_vars = ['wbc', 'neutrophils', 'lymphocytes', 'hgb', 'plt', 'na', 'hco3', 'bun', 'cr', 'lactate', 'procalcitonin']

median_cols = [f'median_{var}' for var in labs_vars]
first_cols = [f'first_{var}' for var in labs_vars]
last_cols = [f'last_{var}' for var in labs_vars]

labs = labs[labs_cols_base + median_cols + first_cols + last_cols]

for col in median_cols + first_cols + last_cols:
    labs[col] = pd.to_numeric(labs[col], errors='coerce')

for var in labs_vars:
    first = labs[f'first_{var}']
    last = labs[f'last_{var}']
    labs[f'delta_{var}'] = last - first
    labs.loc[first.isna() | last.isna(), f'delta_{var}'] = pd.NA

labs_final = labs[labs_cols_base + median_cols + [f'delta_{var}' for var in labs_vars]]
labs_final = labs_final.dropna(subset=median_cols + [f'delta_{var}' for var in labs_vars], how='all')

print(labs_final.info())
````

OMG extremely sparse - Yarab maysabib mashakel ba3den:)
- 13 `median_*` features — good overall coverage
- 12 `delta_*` features — most are extremely sparse or entirely missing (TT_TT).

| Feature                             | Non-null Count | Notes                          |
|-------------------------------------|----------------|--------------------------------|
| `delta_wbc`                         | 647            | Usable, but <10% of rows       |
| `delta_na`, `delta_cr`, `delta_bun`| 100–140        | Marginally usable              |
| `delta_lactate`                     | 7              | Nearly unusable                |
| `delta_procalcitonin`, `delta_hgb`, etc. | 0              | Unusable (no data available)   |


### What This Implies

- **`median_*` features are solid** — they cover most rows and summarize central trends well.

- **Most `delta_*` features are too sparse to be useful**:
  - Using them in modeling may either introduce noise or cause instability 😭.
  - Features with **0 non-null values** should be **dropped**.
  - For **low-count features** (e.g., <1–5% non-null), it may be better to drop them:
    - But i will use **models robust to missingness** (e.g., XGBoost with `missing` parameter).
    - So i will drop columns with 0 non-null values or <1% non-null


````python
threshold = 0.01  # 1%

labs_filtered = labs_final.dropna(axis=1, thresh=int(len(labs) * threshold))

print(f"Original shape: {labs_final.shape}")
print(f"Filtered shape: {labs_filtered.shape}")
````

## cultures_comorbidity.parquet file preparing

````python
import re

def map_comorbidity_to_group(label: str) -> str:
    s = label.lower()

    # diabetes
    if "diabetes" in s:
        return "has_diabetes"

    # heart failure / CHF
    if ("congestive heart failure" in s) or ("heart failure" in s) or ("chf" in s):
        return "has_chf"

    # kidney disease
    if ("chronic kidney" in s) or ("ckd" in s) or ("renal failure" in s) or ("nephritis" in s) or ("nephrosis" in s) or ("renal sclerosis" in s):
        return "has_ckd"

    # cancer / malignancy
    if any(k in s for k in ["cancer", "malignant", "neoplasm", "leukemia", "lymphoma", "myeloma", "metastatic", "sarcoma", "tumor"]):
        return "has_cancer"

    # transplant
    if "transplant" in s:
        return "has_transplant"

    # immunosuppression / immunity disorders / HIV
    if any(k in s for k in ["immunity disorder", "immunodeficiency", "hiv", "aids", "immunosuppress"]):
        return "has_immunosuppression"

    # COPD / chronic pulmonary
    if any(k in s for k in ["copd", "emphysema", "chronic bronchitis", "bronchiectasis", "chronic pulmonary disease"]):
        return "has_copd"

    # liver disease
    if any(k in s for k in ["cirrhosis", "hepatic", "hepatitis", "liver disease", "liver failure"]):
        return "has_liver_disease"

    # dementia
    if any(k in s for k in ["dementia", "alzheimer", "neurocognitive"]):
        return "has_dementia"

    # stroke / CVA / cerebrovascular
    if any(k in s for k in ["stroke", "cva", "cerebral infarction", "cerebrovascular", "transient cerebral ischemia"]):
        return "has_stroke"

    return "has_other_comorbid"
````

````python
keys = ["anon_id","pat_enc_csn_id_coded","order_proc_id_coded","order_time_jittered_utc"]

active = como[
    (como["comorbidity_component_start_days_culture"] >= 0) &
    ((como["comorbidity_component_end_days_culture"].isna()) | (como["comorbidity_component_end_days_culture"] >= 0))
].copy()

active["group"] = active["comorbidity_component"].apply(map_comorbidity_to_group)

grouped_wide = (
    pd.crosstab(index=[active[k] for k in keys], columns=active["group"])
      .clip(upper=1).astype("int8").reset_index()
)

# ensure all expected columns exist
expected = [
    "has_diabetes","has_chf","has_ckd","has_cancer","has_transplant",
    "has_immunosuppression","has_copd","has_liver_disease","has_dementia",
    "has_stroke","has_other_comorbid"
]
for c in expected:
    if c not in grouped_wide.columns:
        grouped_wide[c] = 0

grouped_wide = grouped_wide[keys + expected]
print(grouped_wide.shape)
````

### Utility: evaluate probabilities + threshold optimization

````python
def pick_threshold(y_true: np.ndarray, y_prob: np.ndarray, metric: str = "f1") -> float:
    """
    Choose a threshold on probabilities using the specified metric on y_true.
    metric: 'f1' (default) or 'youden' (sensitivity+specificity-1)
    """
    thresholds = np.linspace(0.01, 0.99, 99)
    best_t, best_val = 0.5, -np.inf

    for t in thresholds:
        y_pred = (y_prob >= t).astype(int)

        if metric == "f1":
            val = f1_score(y_true, y_pred, zero_division=0)
        elif metric == "youden":
            tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
            sens = tp / (tp + fn) if (tp + fn) else 0.0
            spec = tn / (tn + fp) if (tn + fp) else 0.0
            val = sens + spec - 1
        else:
            raise ValueError("metric must be 'f1' or 'youden'")

        if val > best_val:
            best_val = val
            best_t = t

    return float(best_t)


def eval_probs(y_true: np.ndarray, y_prob: np.ndarray, threshold: float = 0.5) -> dict:
    y_pred = (y_prob >= threshold).astype(int)
    return {
        "roc_auc": roc_auc_score(y_true, y_prob),
        "pr_auc": average_precision_score(y_true, y_prob),
        "brier": brier_score_loss(y_true, y_prob),
        "f1": f1_score(y_true, y_pred, zero_division=0),
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred, zero_division=0),
        "threshold": threshold
    }
````

### Utility: group-aware CV scoring function (for Optuna)

````python
def cv_score_model(make_model_fn, X, y, groups, cv, use_metric="pr_auc") -> float:
    scores = []

    for tr_idx, va_idx in cv.split(X, y, groups=groups):
        X_tr, y_tr = X.iloc[tr_idx], y.iloc[tr_idx]
        X_va, y_va = X.iloc[va_idx], y.iloc[va_idx]

        model = make_model_fn()
        model.fit(X_tr, y_tr)

        # Probabilities
        y_va_prob = model.predict_proba(X_va)[:, 1]

        if use_metric == "pr_auc":
            score = average_precision_score(y_va, y_va_prob)
        elif use_metric == "roc_auc":
            score = roc_auc_score(y_va, y_va_prob)
        else:
            raise ValueError("use_metric must be 'pr_auc' or 'roc_auc'")

        scores.append(score)

    return float(np.mean(scores))
````

````python
import numpy as np
import pandas as pd

def make_model_ready(X: pd.DataFrame) -> pd.DataFrame:
    X = X.copy()

    # 1) Replace pandas NA (NAType) with np.nan
    X = X.replace({pd.NA: np.nan})

    # 2) Convert nullable integer/float extension dtypes to numpy floats
    # (XGBoost handles np.nan in float columns)
    for col in X.columns:
        if pd.api.types.is_integer_dtype(X[col]) and str(X[col].dtype).startswith("Int"):
            X[col] = X[col].astype("float32")
        elif pd.api.types.is_bool_dtype(X[col]) and str(X[col].dtype) == "boolean":
            # boolean extension dtype -> float (or int)
            X[col] = X[col].astype("float32")
        elif pd.api.types.is_string_dtype(X[col]) or str(X[col].dtype) == "string":
            # You should not have strings in X now; if you do, drop/encode them
            raise ValueError(f"String column still in X: {col}")

    # 3) Finally, ensure everything numeric
    bad = X.select_dtypes(exclude=["number"]).columns.tolist()
    if bad:
        raise ValueError(f"Non-numeric columns remain: {bad}")

    return X
````

````python
import numpy as np
from sklearn.metrics import roc_auc_score, average_precision_score, brier_score_loss

def bootstrap_ci(y_true, y_prob, metric_func, n_bootstrap=1000, random_state=42):
    rng = np.random.RandomState(random_state)
    scores = []

    n = len(y_true)

    for _ in range(n_bootstrap):
        indices = rng.randint(0, n, n)
        
        if len(np.unique(y_true.iloc[indices])) < 2:
            continue  # skip if only one class sampled
        
        score = metric_func(y_true.iloc[indices], y_prob[indices])
        scores.append(score)

    lower = np.percentile(scores, 2.5)
    upper = np.percentile(scores, 97.5)

    return np.mean(scores), lower, upper
````

````python
from sklearn.metrics import roc_auc_score, average_precision_score, brier_score_loss

def patient_bootstrap_ci(
    df_eval: pd.DataFrame,
    patient_col: str = "anon_id",
    y_col: str = "y",
    p_col: str = "p",
    n_bootstrap: int = 1000,
    random_state: int = 42,
):
    rng = np.random.default_rng(random_state)

    patients = df_eval[patient_col].astype(str).unique()
    n_patients = len(patients)

    roc_scores, pr_scores, brier_scores = [], [], []

    for _ in range(n_bootstrap):
        # sample patients WITH replacement
        sampled_patients = rng.choice(patients, size=n_patients, replace=True)

        # collect all rows for sampled patients (duplicates allowed)
        # easiest: concatenate per-patient chunks
        chunks = [df_eval[df_eval[patient_col].astype(str) == pid] for pid in sampled_patients]
        boot = pd.concat(chunks, axis=0, ignore_index=True)

        yb = boot[y_col].to_numpy()
        pb = boot[p_col].to_numpy()

        # If bootstrap sample accidentally has only one class, skip it
        if np.unique(yb).size < 2:
            continue

        roc_scores.append(roc_auc_score(yb, pb))
        pr_scores.append(average_precision_score(yb, pb))
        brier_scores.append(brier_score_loss(yb, pb))

    def ci(arr):
        arr = np.asarray(arr, dtype=float)
        return arr.mean(), np.percentile(arr, 2.5), np.percentile(arr, 97.5)

    return {
        "roc_auc": ci(roc_scores),
        "pr_auc": ci(pr_scores),
        "brier": ci(brier_scores),
        "n_effective_bootstraps": (len(roc_scores), len(pr_scores), len(brier_scores)),
        "n_patients": n_patients,
        "n_rows": len(df_eval),
    }
````

### Fit best model on full train, then calibrate (group-aware)

Calibration should be done only using training data, and ideally using CV.
We’ll use `CalibratedClassifierCV` with GroupKFold, which calibrates out-of-fold.



### Helper: train + calibrate

````python
def fit_and_calibrate(base_model, X_train, y_train, groups_train, cv):
    """
    Returns a calibrated model (isotonic by default for best calibration; use sigmoid if you want lighter).
    """
    calibrated = CalibratedClassifierCV(
        estimator=base_model,
        method="isotonic",  # 'sigmoid' is faster; isotonic can be better but heavier
        cv=cv
    )
    calibrated.fit(X_train, y_train, groups=groups_train)
    return calibrated
````

````python
# TODO: (Repeat for other models.)
def bootstrap_ci_by_patient(
    cal_model, X_test, y_test, g_test, threshold,
    n_boot=300, seed=42
):
    rng = np.random.default_rng(seed)
    patients = np.unique(g_test)

    metrics_list = []
    for _ in range(n_boot):
        sampled_patients = rng.choice(patients, size=len(patients), replace=True)
        mask = np.isin(g_test, sampled_patients)

        y_prob = cal_model.predict_proba(X_test[mask])[:, 1]
        m = eval_probs(y_test[mask].to_numpy(), y_prob, threshold=threshold)
        metrics_list.append(m)

    mdf = pd.DataFrame(metrics_list)
    ci = {}
    for col in ["roc_auc", "pr_auc", "brier", "f1", "precision", "recall"]:
        ci[col] = {
            "mean": float(mdf[col].mean()),
            "lo": float(mdf[col].quantile(0.025)),
            "hi": float(mdf[col].quantile(0.975)),
        }
    return ci

ci_xgb = bootstrap_ci_by_patient(xgb_cal, X_test, y_test, g_test, t_xgb, n_boot=200)
ci_xgb
````
