# Read all files

i will do this because i need to know the data types for all columns 

```python
import pandas as pd
import glob
import os

folder_path = "doi_10_5061_dryad_jq2bvq8kp__v20250411"

all_files = glob.glob(os.path.join(folder_path, "*.csv"))
all_df_files = {}
all_features = []
for file in all_files:
    df = pd.read_csv(file, nrows = 5)
    label = os.path.splitext(os.path.basename(file))[0]
    all_df_files[label] = df
    all_features.extend(df.columns.tolist())

unique_features = set(all_features)

print(f'features = {unique_features}')
print(f'No of features = {len(unique_features)}')

for file, df in all_df_files.items():
    print(f'file: {file}')
    print(df.dtypes)
    
```

```python
linked_features = ['anon_id', 'pat_enc_csn_id_coded', 'order_proc_id_coded', 'order_time_jittered_utc']
base_file = all_df_files['microbiology_cultures_cohort']
for df in list(all_df_files.keys())[1:]:
    current_df = all_df_files[df]
    if all(feature in base_file.columns for feature in linked_features) and \
       all(feature in current_df.columns for feature in linked_features):
        base_file = pd.merge(base_file, all_df_files[df], on = linked_features, how="inner")

print(base_file)
```
